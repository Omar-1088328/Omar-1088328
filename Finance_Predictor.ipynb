# Data Collection

First Step which is most important, Data collection

I will be using the yfinance API to get real-time stock data from the stock market. I will satart off small with scraping only 1 stock, then building a pipeline where i can do some feature engineering and create some labels, then i will add some more stock.

For the first stock, i will start off with Apple (AAPL)

```python
data = yf.download('AAPL', period = '1y', interval = '1d')

df = pd.DataFrame(data)

data.reset_index(inplace=True)

df.head(10)
```

It looks like the columns are grouped under a heading like "Price â†’ Open, High, Low, Close". lets fix that

```python
# If columns are multi-index
if isinstance(df.columns, pd.MultiIndex):
    # Flatten the columns so for Example: 'Price Close' becomes 'Price_Close'
    df.columns = ['_'.join(col).strip() if isinstance(col, tuple) else col for col in df.columns]

# Rename columns to be even simpler
df = df.rename(columns={
    'Price_Open': 'Open',
    'Price_High': 'High',
    'Price_Low': 'Low',
    'Price_Close': 'Close',
    'Price_Volume': 'Volume'
})

# Reset index to make 'Date' column a normal column
df = df.reset_index()

df.head(10)
```
